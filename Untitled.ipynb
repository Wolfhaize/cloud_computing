```python
import matplotlib.pyplot as plt
import numpy as np

# Data
load = np.array([100, 500, 1000, 1500, 2000])

lat_bare = np.array([1.2, 2.5, 4.8, 8.2, 15.4])
err_bare = np.array([0.1, 0.2, 0.4, 0.7, 1.2])

lat_hvm = np.array([1.8, 3.6, 7.4, 12.9, 28.5])
err_hvm = np.array([0.2, 0.4, 0.6, 1.1, 2.4])

lat_trap = np.array([4.5, 12.2, 45, 120, 350])
err_trap = np.array([0.5, 1.8, 5.2, 15, 40])

# Bar positions
x = np.arange(len(load))
width = 0.25

plt.figure(figsize=(10,6))

# Bars with error bars
plt.bar(x - width, lat_bare, width, yerr=err_bare, capsize=5, label='Bare Metal')
plt.bar(x, lat_hvm, width, yerr=err_hvm, capsize=5, label='HVM (Virtio/HW)')
plt.bar(x + width, lat_trap, width, yerr=err_trap, capsize=5, label='Trap & Emulate')

# Overlay lines (trend)
plt.plot(x - width, lat_bare, marker='o')
plt.plot(x, lat_hvm, marker='s')
plt.plot(x + width, lat_trap, marker='^')

# Labels
plt.xticks(x, load)
plt.xlabel("Offered Load (requests/sec)")
plt.ylabel("99%-ile Latency (ms)")
plt.title("Latency vs Load (with Std Dev Error Bars)")
plt.legend()
plt.grid(axis='y')

plt.show()
```
